```python
import random
import uuid
import pandas as pd
from faker import Faker

fake = Faker('en_GB')

def generate_data(v):
  if v > 200:
    raise ValueError("number too big")
  record = []
  for i in range(v):
    data = {
        "name": fake.name(),
        "email": fake.email(),
        "phone": fake.phone_number()

    }
    record.append(data)
  return record

generate_data(10)
```

```python
import time


def process_batch(names):
  for i in range(0, len(names), 10):
    yield names[i:i+10]
    time.sleep(1)


```

```python
def transform_data(batches):
  transformed_data =[]
  for batch in batches:
    transform_data = {
        "id": batch["id"],
        "price": batch["price"],
        "apprm_price": batch["price"]*1.1
    }
    transformed_data.append(transform_data)
  return transformed_data
```

```python
data = generate_data(100)


def main():
  whole_data = []
  for batch in process_batch(data):
    result = transform_data(batch)
    whole_data.extend(result)
  df = pd.DataFrame(whole_data)
  df.to_csv("new_transformed_data.csv", index=False)
  display(df)

main()
```
